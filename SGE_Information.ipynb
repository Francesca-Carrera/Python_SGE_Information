##### FUNZIONALITÀ DI BASE.

**1. Visualizzazione di tutti gli elementi** - stampare a schermo tutti gli elementi registrati nella lista. <br>
<br>
- CONTROLLO PRELIMINARE: <br>
Controllo che *articoli_2020_Internazionale* sia un dizionario <br>
e visualizzo un messaggio di errore in caso contrario, terminando così la funzione. <br>
<br>
- VERIFICA DELLE CHIAVI PRINCIPALI: <br>
Per garantire che tutte le chiavi (categorie degli articoli) siano presenti nel dizionario, le inserisco in una lista di confronto. <br>
Utilizzando la funzione `get`, che permette di cercare direttamente nel dizionario e gestire i valori mancanti, <br>
posso identificare e visualizzare le chiavi assenti. In caso di valore mancante, visualizzo un messaggio di errore, <br> 
indicando quale chiave non è presente. <br>
<br>
- VERIFICA DELLA STRUTTURA DEI VALORI: <br>
Una volta verificata la correttezza del dizionario e delle sue chiavi principali, controllo che ogni valore sia una lista <br>
e che ogni elemento in essa sia un dizionario, fornendo messaggi di errore appropriati in caso di problemi <br> 
e verifico infine che ogni dizionario contenga le chiavi necessarie.

```python
def visualizza_articoli() -> None:
    
    # Controllo che articoli_2020_Internazionale sia un dizionario.
    if not isinstance(articoli_2020_Internazionale, dict):
        print("Errore: articoli_2020_Internazionale non è un dizionario.")
        return
     
    # Elementi di una lista, a rappresentazione delle chiavi, che devono essere presenti nel dizionario.   
    categorie_attese = ["Ambiente", "Coronavirus", "Cultura", "Disuguaglianze", "Femminismo",
                        "Italia", "Lavoro", "Migranti", "Mondo", "Relazioni e società", "Scuola"]
        
    # Verifica che gli elementi della lista siano presenti come chiavi nel dizionario.
    for categoria_attesa in categorie_attese:
        articoli = articoli_2020_Internazionale.get(categoria_attesa, None)
        
        # Se la chiave non è presente nel dizionario.     
        if articoli is None:
            print(f"Errore, la categoria: \"{categoria_attesa}\" non è presente.")
            print(f"{"-" * len("Errore")}\n")
            continue
        
        # Controllo che ogni valore del dizionario sia una lista.
        if not isinstance(articoli, list):
            print(f"Errore: la categoria \"{categoria_attesa}\" non dispone di articoli.")
            print(f"{"-" * len("Errore")}\n") # Linea separatrice.
            continue
        
        # Se la chiave è presente nel dizionario. 
        print(f"{categoria_attesa}:")
        print(f"{"-" * len(categoria_attesa)}\n") # Linea separatrice.
        
        # Se il valore del dizionario è una lista.    
        for articolo in articoli:
            # Verifica che ogni elemento della lista sia un dizionario.
            if not isinstance(articolo, dict):
                print(f"Errore: un elemento della lista nella categoria '{str(categoria_attesa)}' non è un dizionario.\n")
                continue
            
            # Se l'elemento della lista è un dizionario si controlla che per ognuno 
            # siano presenti le chiavi "Posizionamento" e "Titolo".
            posizionamento = articolo.get("Posizionamento", None)
            titolo = articolo.get("Titolo", None)
            
            if posizionamento is None and titolo is None:
                print(" Errore, posizionamento dell'articolo non presente.") 
                print(" Errore, titolo dell'articolo non presente. \n") 
                continue
            if posizionamento is None:
                print(" Errore, posizionamento dell'articolo non presente.")
                print(f" Titolo: {str(titolo)}\n") # articolo['Titolo']  
                continue
            if titolo is None:
                print(f" Posizionamento: {str(posizionamento)}") # articolo['Posizionamento']
                print(" Errore, titolo dell'articolo non presente. \n")     
                continue
            
            print(f" Posizionamento: {str(posizionamento)}") # articolo['Posizionamento']      
            print(f" Titolo: {str(titolo)}\n") # articolo['Titolo']  
            
        print(f"{"=" * 100}\n")  # Separatore delle categorie di articoli.
```
